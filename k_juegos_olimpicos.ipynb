Lo siguiente es para calcular el número de ciudades donde se han realizado los juegos.

```python
num_ciudades = df['City'].value_counts().shape
print(num_ciudades)
```

Lo siguiente es para saber la proporción de hombres y mujeres en todos los juegos.

```python
proporcion = df['Sex'].value_counts() / len(df["Sex"])
print(proporcion)
```

```python
ano_primera_mujer = df[df['Sex'] == 'F']['Year'].min()
print(ano_primera_mujer)
```

```python
atletas_grandes = df[(df['Height'] > 190.0) & (df['Weight'] > 100.0)]
print(atletas_grandes)
```

```python
ano_primer_hombre = df[df['Sex'] == 'M']['Year'].min()
print(ano_primer_hombre)
```

```python
hombre_menor_edad = df[df['Sex'] == 'M']['Age'].min()
print(hombre_menor_edad)
```

```python
anos_totales = df['Year'].unique()
print(np.sort(anos_totales))
```

```python
edad_promedio=df['Age'].mean()
print(edad_promedio)
```

```python
datos_oros = df[(df['Age'] < 18) & (df['Medal'] == 'Gold')]
datos_oros.head(10)
```

```python
data_team = df[df['Team'] == 'Chile']
data_team.head(15)
```
